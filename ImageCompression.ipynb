```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
img = plt.imread('pitt.jpg')
img_size = img.shape
```

```python
imgAs2DArr = img.reshape(img_size[0] * img_size[1], img_size[2])

n_colors10 = 10
km10 = KMeans(n_clusters = n_colors10).fit(imgAs2DArr)
less_colors10 = km10.cluster_centers_[km10.labels_]
less_colors10 = np.clip(less_colors10.astype('uint8'), 0, 255)
less_colors10 = less_colors10.reshape(img_size[0], img_size[1], img_size[2])

n_colors6 = 6
km6 = KMeans(n_clusters = n_colors6).fit(imgAs2DArr)
less_colors6 = km6.cluster_centers_[km6.labels_]
less_colors6 = np.clip(less_colors6.astype('uint8'), 0, 255)
less_colors6 = less_colors6.reshape(img_size[0], img_size[1], img_size[2])

n_colors2 = 2
km2 = KMeans(n_clusters = n_colors2).fit(imgAs2DArr)
less_colors2 = km2.cluster_centers_[km2.labels_]
less_colors2 = np.clip(less_colors2.astype('uint8'), 0, 255)
less_colors2 = less_colors2.reshape(img_size[0], img_size[1], img_size[2])
```

```python
fig, ax = plt.subplots(1,4, figsize = (12,8))
ax[0].imshow(img)
ax[0].set_title('Original Image')
ax[1].imshow(less_colors10)
ax[1].set_title('10 colors')
ax[2].imshow(less_colors6)
ax[2].set_title('6 Colors')
ax[3].imshow(less_colors2)
ax[3].set_title('2 Colors')
for ax in fig.axes:
    ax.axis('off')
    plt.tight_layout();
```
